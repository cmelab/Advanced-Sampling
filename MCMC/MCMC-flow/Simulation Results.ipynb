```python
import signac
project = signac.get_project()
```

```python
max_trans_list = []
ar_list = []
for job in project:
    max_trans = job.sp['max_trans'][1]
    accep_ratio = job.doc['acceptance_ratio'][1]
    if max_trans not in max_trans_list:
        max_trans_list.append(max_trans)
    if accep_ratio not in ar_list:
        ar_list.append(accep_ratio)
max_trans_list.sort
ar_list.sort
```

```python
energy = []
for job in project:
    E = job.doc['energy'][1]
    if E not in energy:
        energy.append(E)
energy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(max_trans_list, ar_list,'o')
plt.show()
```

```python
plt.plot(energy,'o')
plt.show()
```
